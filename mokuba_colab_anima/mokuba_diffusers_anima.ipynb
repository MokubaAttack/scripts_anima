## run

```python
#@title text2image

#@markdown ##ckpt file
base_safe="ckpt.safetensors" #@param {type:"string"}

#@markdown ##lora files
loras="lora1, lora2" #@param {type:"string"}
if loras=="":
    loras=[]
else:
    loras=loras.split(",")
#@markdown ##lora weights
lora_weights="1, 1" #@param {type:"string"}
if lora_weights=="":
    lora_weights=[]
else:
    lora_weights=lora_weights.split(",")
    for i in range(len(lora_weights)):
        lora_weights[i]=float(lora_weights[i])

#@markdown ##prompt
prompt = "" #@param {type:"string"}
#@markdown ##negative prompt
n_prompt = "" #@param {type:"string"}

#@markdown ##output size
x=1024# @param {"type":"integer"}
y=1024# @param {"type":"integer"}
#@markdown ##Hires upscale
up=1.5# @param {"type":"number"}
#@markdown ##program type
prog_ver="1:hires.fix" #@param ["0:nomal", "1:hires.fix"] {allow-input: false}
prog_ver=int(prog_ver.split(":")[0])

#@markdown ##number of output image
pic_number=10# @param {"type":"integer"}
#@markdown ##guidance_scale
gs=4# @param {"type":"number"}
#@markdown ##num_inference_steps
step=30# @param {"type":"integer"}
#@markdown ##Hires steps
step2=15# @param {"type":"integer"}
#@markdown ##denoising_strength
ss=0.4# @param {"type":"number"}

#@markdown ##seed
seed="0"# @param {"type":"string"}
if "," in seed:
    seed=seed.split(",")

#@markdown ##interpolation method
Interpolation="1:NEAREST" #@param ["1:NEAREST","2:BOX","3:BILINEAR","4:HAMMING","5:BICUBIC","6:LANCZOS"] {allow-input: true}
if Interpolation in ["1:NEAREST","2:BOX","3:BILINEAR","4:HAMMING","5:BICUBIC","6:LANCZOS"]:
    Interpolation=int(Interpolation.split(":")[0])
#@markdown ##Scheduler
sample="flowmatch_euler" #@param ["flowmatch_euler","euler","euler_a_rf","euler_ancestral_rf"] {allow-input: false}
#@markdown ##noise schedule and schedule type
sgm="uniform" #@param ["uniform","beta","simple","normal"] {allow-input: false}

#@markdown ##Output folder
out_folder=""#@param {type:"string"}
#@markdown ##discord webhook url
discord_url=""#@param {type:"string"}

import mokucola

pipe=mokucola.mokuani(
    base_safe=base_safe,
    loras=loras, 
    lora_weights=lora_weights, 
    prompt=prompt, 
    n_prompt=n_prompt,  
    pic_number=pic_number, 
    gs=gs,  
    step=step, 
    step2=step2,
    ss=ss, 
    Interpolation=Interpolation, 
    sample=sample,
    sgm=sgm,
    seed=seed,
    out_folder=out_folder,
    x=x,
    y=y,
    up=up,
    p=None,
    url=discord_url,
    dtype="f32",
    dev="cuda",
    ser="colab",
    del_pipe=True,
    si=True,
    mode=prog_ver
)
```
